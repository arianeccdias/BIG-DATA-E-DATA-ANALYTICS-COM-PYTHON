```python
import numpy as np
fig, ax = plt.subplots()

Mobilidade = ['Abadia de Goiás','Abadiânia','Abaeté']
counts = [2158, 2838, 6928]
bar_labels = ['red', 'blue', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:orange']

ax.bar(Mobilidade, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('Mobilidade')
ax.set_title('Carros por cidades de Goiás')
ax.legend(title='Carros em 3 cidades de Goiás')

plt.show()
```
