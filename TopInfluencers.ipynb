Importando biblioteca de gráficos

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
#definindo os dados
data = [485.2,370.7,363.9,348.8,339.4]
labels =['\ncristiano', 'kyliejenner', '\nleomessi', '\nselenagomez', '\ntherock']

#definindo a paleta de cores
colors = sns.color_palette('pastel')[0:5]
#colors sns.color_palette('tab10')

#cria o gráfico
plt.pie(data, labels = labels, colors = colors, autopct= '%.0f%%')
plt.show()
```

```python
import numpy as np
fig, ax = plt.subplots()

Top_Influencers = ['\ncristiano','kyliejenner','\nleomessi','\nselenagomez']
counts = [485.2,370.7,363.9,348.8]
bar_labels = ['red', 'blue', 'red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

ax.bar(Top_Influencers, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('Top_Influencers')
ax.set_title('Influencers by kind and color')
ax.legend(title='Influencers')

plt.show()
```
